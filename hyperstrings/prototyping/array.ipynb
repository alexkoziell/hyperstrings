```python
from hyperstrings.hypergraph.drawable import DrawableHypergraph
```

```python
hypergraph = DrawableHypergraph()
for s in 'fgh':
    hypergraph.add_hyperedge(s)
for s in 'ABCDEABC':
    hypergraph.add_vertex(s)

hypergraph.inputs = [0, 1]
hypergraph.outputs = [3, 4, 5, 6]

hypergraph.connect_source(0, 0, 0); hypergraph.connect_source(1, 0, 1); hypergraph.connect_target(2, 0, 0); hypergraph.connect_target(7, 0, 1)
hypergraph.connect_source(2, 1, 0); hypergraph.connect_target(3, 1, 0); hypergraph.connect_target(4, 1, 1); hypergraph.connect_source(7, 1, 1)
hypergraph.connect_source(2, 2, 0); hypergraph.connect_target(5, 2, 0); hypergraph.connect_target(6, 2, 1); hypergraph.connect_source(7, 2, 1)

hypergraph.draw_matplotlib()
```
