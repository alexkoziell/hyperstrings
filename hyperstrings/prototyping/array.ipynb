```python
from hyperstrings.hypergraph.drawable import DrawableHypergraph

f = DrawableHypergraph.generator('AB', 'C', 'f')
g = DrawableHypergraph.generator('C', 'AB', 'g')

h = ((f @ g) >> (g @ f))
h.connect_source(4, 3, 2)
h.connect_source(4, 2, 1)
h.connect_target(4, 0, 1)
h.connect_target(4, 1, 2)
h.connect_source(2, 0, 2)

h.draw_matplotlib()
```

```python
h.connect_source(5, 3, 3)
h.connect_target(2, 0, 2)
h.connect_source(4, 1, 1)
h.draw_matplotlib()
```

```python
h_expl = h.explicit_spiders()
h_expl.draw_matplotlib()
```
