```python
from hyperstrings.hypergraph.adjacencytensor.hypergraph import Hypergraph

f = Hypergraph.generator('AB', 'C', 'f')
g = Hypergraph.generator('C', 'AB', 'g')

h = ((f @ g) >> (g @ f))
h.connect_source(4, 3, 2)
h.connect_source(4, 2, 1)
h.connect_target(4, 0, 1)
h.connect_target(4, 1, 2)
h.connect_source(2, 0, 2)

h.draw_matplotlib()
```
